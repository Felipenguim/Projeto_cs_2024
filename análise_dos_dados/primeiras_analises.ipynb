```python
df['KAST'] = df['KAST'].str.rstrip('%').astype('float') / 100.0
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = ['Rating 2.0', 'DPR', 'KPR', 'Impact', 'ADR', 'KAST']


fig, axes = plt.subplots(2, 3, figsize=(15, 10))

colors_top = ['blue', 'red', 'green']
colors_bottom = ['purple', 'orange', 'brown']

# Iterar sobre as colunas e os eixos do grid
for i, (ax, col) in enumerate(zip(axes.flatten(), columns)):
    color = colors_top[i] if i < 3 else colors_bottom[i - 3]
    sns.histplot(df[col], kde=True, ax=ax, color=color)
    ax.set_title(f'Distribuição de {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
highlight_df = pd.DataFrame()
for col in columns:
    if col == 'DPR':
        best_players = df.nlargest(5, col)[['Nick', col]].assign(Type='Pior')
        worst_players = df.nsmallest(5, col)[['Nick', col]].assign(Type='Melhor')
        highlight_df = pd.concat([highlight_df, best_players, worst_players])
    else:
        best_players = df.nlargest(5, col)[['Nick', col]].assign(Type='Melhor')
        worst_players = df.nsmallest(5, col)[['Nick', col]].assign(Type='Pior')
        highlight_df = pd.concat([highlight_df, best_players, worst_players])

# Plotar os resultados
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
palette = {'Melhor': 'blue', 'Pior': 'red'}

for i, col in enumerate(columns):
    ax = axes.flatten()[i]
    sns.barplot(data=highlight_df[highlight_df[col].notnull()], x=col, y='Nick', hue='Type', dodge=False, ax=ax, palette=palette)
    ax.set_title(f'Top 5 e Bottom 5 de {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Jogador')

plt.tight_layout()
plt.show()
```

```python
def melhor_pior(columns, df):
    highlight_df = pd.DataFrame()
    for col in columns:
        best_players = df.nlargest(5, col)[['Nick', col]].assign(Type='Melhor')
        worst_players = df.nsmallest(5, col)[['Nick', col]].assign(Type='Pior')
        highlight_df = pd.concat([highlight_df, best_players, worst_players])

    # Plotar os resultados
    n = len(columns)
    if(n%2 != 0):
        n += 1
    fig, axes = plt.subplots(2, int(n/2), figsize=(15, 10))
    palette = {'Melhor': 'blue', 'Pior': 'red'}

    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        sns.barplot(data=highlight_df[highlight_df[col].notnull()], x=col, y='Nick', hue='Type', dodge=False, ax=ax, palette=palette)
        ax.set_title(f'Top 5 e Bottom 5 de {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Jogador')

    plt.tight_layout()
    plt.show()
```
