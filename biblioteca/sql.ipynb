```python
import pandas as pd
import re
```

```python
df = pd.read_csv('stats_palyers_2024.csv')
df
```

```python
df = df.drop(columns=["0.1", "0"])
df
```

```python
def limpar_nomes_colunas(colunas):
    colunas_limpas = []
    for col in colunas:
        # Remover caracteres que não sejam letras, números ou '_'
        coluna_limpa = re.sub(r"[^\w]", "", col)
        colunas_limpas.append(coluna_limpa)
    return colunas_limpas

# Aplicar a limpeza nos nomes das colunas
df.columns = limpar_nomes_colunas(df.columns)
```

```python
def gerar_create_table(df, nome_tabela):
    sql = f"CREATE TABLE {nome_tabela} (\n"
    for col, dtype in zip(df.columns, df.dtypes):
        if dtype == 'int64' or dtype == 'float64':
            tipo_sql = 'NUMBER (5, 2)'
        elif dtype == 'object':
            tipo_sql = 'VARCHAR2(255)'
        else:
            tipo_sql = 'VARCHAR2(255)'
        sql += f"    {col} {tipo_sql},\n"
    sql = sql.rstrip(',\n') + "\n);"
    return sql

print(gerar_create_table(df, 'minha_tabela'))
```
